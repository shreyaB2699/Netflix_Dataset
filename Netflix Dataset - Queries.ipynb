```python
# FUNCTION TO CHECK MEDIA FROM ANY COUNTRY

def check_media_count(coun):
    a = %sql SELECT COUNT(*) FROM NETFLIX_TABLE WHERE COUNTRY LIKE '%{coun}%'
    b= %sql SELECT COUNT(*) FROM NETFLIX_TABLE WHERE COUNTRY = :coun
    print(f'MEDIA FROM {coun}')
    print(a)
    print(f'MEDIA FROM {coun} only')
    print(b)

    
check_media_count('United States')
```

```python
# Function for MOVIES/TV Shows with any actor

def show_actor_work(actor_name):
    a = %sql SELECT TITLE FROM NETFLIX_TABLE WHERE TYPE = 'Movie' AND CAST LIKE '%{actor_name}%'
    print(f'Works of actor {actor_name} are: ')
    print(a)
    
show_actor_work('Kristen Stewart')
```

```python
# FUNCTION FOR DIRECTOR's WORK

def direc_work(director_name):
    a = %sql SELECT TITLE FROM NETFLIX_TABLE WHERE DIRECTOR LIKE '%{director_name}%'
    print(f'Works of Director {director_name} are: ')
    print(a)

direc_work('Quentin Tarantino')
direc_work('Christopher Nolan')
direc_work('Steven Spielberg')
```

```python
# DATE MANIPULATION

%sql SELECT TITLE FROM NETFLIX_TABLE WHERE RELEASE_YEAR > '2014' LIMIT 50
```

```python
# FUNCTION TO PRINT THE MEDIA FOR A PARTICULAR RELEASE YEAR

def media_release_year(year):
    a = %sql SELECT TITLE FROM NETFLIX_TABLE WHERE RELEASE_YEAR = '{year}'
    print(f'The titles of shows in year {year} are')
    print(a)
    
media_release_year('2020')
```

```python
# GETTING TITLES FOR A SPECIFIC RATING

def media_rating(rating):
    a = %sql SELECT TITLE, TYPE, DURATION FROM NETFLIX_TABLE WHERE RATING = '{rating}'
    print(f'The titles of shows for rating {rating} are')
    print(a)
    
media_rating('R')
```

```python
# FUNCTION TO GET TITLES FOR SPECIFIC GENRE

def media_genre(genre):
    a = %sql SELECT TITLE, TYPE, DURATION FROM NETFLIX_TABLE WHERE LISTED_IN LIKE '%{genre}%'
    print(f'The titles of shows for genre {genre} are')
    print(a)
    
media_genre('Dramas')
```

```python
media_genre('Sci-Fi') # SCI-FI GENRE
```

```python
# GET DESCRIPTION FOR A MOVIE
def get_description(movie_name):
    a = %sql SELECT TITLE,DESCRIPTION FROM NETFLIX_TABLE WHERE TITLE LIKE '%{movie_name}%'
    print(f'Movie description for title matching {movie_name} is as follows:')
    print(a)
```
